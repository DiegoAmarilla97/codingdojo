```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('2015.csv')
df.head()
```

```python
plt.hist(df['Happiness Score'],color = 'darkred', edgecolor='black', alpha=0.8)
plt.title('Happiness Score')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()
```

```python
plt.hist(df['Economy (GDP per Capita)'],color = 'darkgreen', edgecolor='black', alpha=0.8)
plt.title('Economy (GDP per Capita)')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()
```

```python
plt.hist(df['Freedom'],color = 'blue', edgecolor='black', alpha=0.8)
plt.title('Freedom')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()
```

```python
plt.hist(df['Health (Life Expectancy)'],color = 'red', edgecolor='black', alpha=0.8)
plt.title('Health (Life Expectancy)')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()
```

```python
import seaborn as sns

# Configurar el estilo de los grficos
sns.set(style="whitegrid")

# Definir las variables
variables = ['Happiness Score', 'Economy (GDP per Capita)', 'Health (Life Expectancy)', 'Freedom']

# Definir una paleta de colores
colors = ['#FF6F61', '#6B5B95', '#88B04B', '#F7CAC9']

# Crear una figura con subplots para cada variable
fig, axes = plt.subplots(4, 2, figsize=(12, 16))

for i, var in enumerate(variables):
    # Histograma
    sns.histplot(df[var], bins=20, color=colors[i], edgecolor='black', alpha=0.7, ax=axes[i, 0])
    axes[i, 0].set_title(f'Histograma de {var}', fontsize=14)
    
    # Boxplot
    sns.boxplot(x=df[var], color=colors[i], ax=axes[i, 1], fliersize=5, linewidth=1.5, whis=1.5)
    axes[i, 1].set_title(f'Boxplot de {var}', fontsize=14)

# Ajustar la separacin entre los grficos
plt.tight_layout()
plt.show()
```
