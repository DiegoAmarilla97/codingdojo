```python
#Crea nuevas columnas basadas en operaciones con las columnas existentes (por ejemplo, calcular ingresos a partir de ventas y precios).
meanVotes = df['numVotes'].mean()
df['ponVotes'] = df['averageRating']*df['numVotes']/meanVotes 

print(df.sort_values('ponVotes',ascending=False).head())
```
